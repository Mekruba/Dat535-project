```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, split, explode, regexp_replace, to_date, mean

spark = SparkSession.builder.appName("DataCleaning").getOrCreate()

filepath = "project/anime-dataset-2023.csv"

# Define the schema to match the data structure, forcing each field to be read as String
schema = "anime_id STRING, Name STRING, `English name` STRING, `Other name` STRING, Score STRING, Genres STRING, Synopsis STRING, Type STRING, Episodes STRING, Aired STRING, Premiered STRING, Status STRING, Producers STRING, Licensors STRING, Studios STRING, Source STRING, Duration STRING, Rating STRING, Rank STRING, Popularity STRING, Favorites STRING, `Scored By` STRING, Members STRING, `Image URL` STRING"

anime_df = spark.read.csv(filepath, sep=",", schema=schema, quote="\"", escape="\"", header=True, multiLine=True)

df_cleaned = anime_df.withColumn("Synopsis", regexp_replace(col("Synopsis"), "\n", " ")) \
            .withColumn("Genres", regexp_replace(col("Genres"), ", ", ",").alias("Genres"))

# Display cleaned data
df_cleaned.show(truncate=False)
```
